## Objetivo
Desarrollar un programa que extraiga informacion de una base NoSQL bajo las siguientes condiciones:



1.   El programa inicia preguntando la cantidad de columnas que se va a extraer(1p).
2.   El usuario ingresa uno a uno los campos a consultar(1p).
3.   El programa explora cada uno de los sub diccionarios y solo extrae informacion de los diccionarios que tengan al menos 1 valor de los solicitados, los campos solicitados que no esteen presentes deberan aparecer registrados como un gion '-'(6p).
4.   El programa muestra en pantalla los datos encontrados(2p).


```python
def extraer_informacion(base_de_datos):
    num_columnas = int(input("Ingrese la cantidad de columnas a extraer: "))

    campos_solicitados = []
    for i in range(num_columnas):
        campo = input(f"Ingrese el nombre del campo {i + 1}: ")
        campos_solicitados.append(campo)

    resultados = []

    for documento in base_de_datos:
        resultado_documento = {}
        for campo in campos_solicitados:
            valor = documento.get(campo, '-')
            resultado_documento[campo] = valor

        if any(valor != '-' for valor in resultado_documento.values()):
            resultados.append(resultado_documento)

    print("\nDatos encontrados:")
    for i, resultado in enumerate(resultados, 1):
        print(f"Resultado {i}: {resultado}")

# Base de datos de ejemplo
base_de_datos_ejemplo = [
    {"nombre": "sergio", "apellido": "diaz", "carrera": "economia"},
    {"nombre": "luis", "apellidos": "rios"},
    {"nombre": "noemi", "apellido": "osco", "carrera": "administracion"},
    {"nombre": "juan", "carrera": "derecho"},
    {"nombre": "david", "apellido": "martinez", "carrera": "ingenieria"},
    {"nombre": "maria", "apellido": "romero", "carrera": "economia"},
    {"nombre": "eduardo", "apellido": "sanchez", "carrera": "economia"},
    {"nombre": "emilia", "apellido": "torres", "carrera": "derecho"},
    {"nombre": "rosario", "apellido": "sosa", "carrera": "economia"}
]

# Llamar a la función con la base de datos de ejemplo
extraer_informacion(base_de_datos_ejemplo)

# Bucle para almacenar números hasta que sumen 10
numeros = []
suma_actual = 0

while suma_actual < 10:
    numero = int(input("Ingrese un número: "))
    numeros.append(numero)
    suma_actual = sum(numeros)

print(f"La lista de números es: {numeros}")
```

Crear una funcion que haga 5 preguntas y que imprima el mensaje de aceptar si es que el usuario responde positivamente al menos a 3 de estas preguntas.

Adicionalmente el usuario solo debera ingresar si o no(ya sea en mayusculas o minusculas).

Si no responde correctamente se le debera insistir para que responda si o no. Si tiene 3 o mas respuestas positivas entonces se admite el perfil, caso contrario se le agradece la participacion.

```python
def evaluar_perfil():
    preguntas = [
        "¿Tiene experiencia en programación?",
        "¿Le gusta trabajar en equipo?",
        "¿Está dispuesto/a a aprender nuevas habilidades?",
        "¿Tiene buen manejo de herramientas informáticas?",
        "¿Puede cumplir con plazos de entrega?"
    ]

    respuestas_positivas = 0

    for pregunta in preguntas:
        respuesta = input(f"{pregunta} (responda con 'si' o 'no'): ").lower()

        while respuesta not in ['si', 'no']:
            print("Por favor, responda solo con 'si' o 'no'.")
            respuesta = input(f"{pregunta} (responda con 'si' o 'no'): ").lower()

        if respuesta == 'si':
            respuestas_positivas += 1

    if respuestas_positivas >= 3:
        print("Perfil aceptado. ¡Bienvenido/a!")
    else:
        print("Gracias por participar. No cumplió con los requisitos mínimos.")

# Llamar a la función para evaluar el perfil
evaluar_perfil()
```

# Pregunta 3(5 puntos)

Genere un codigo que le pregunte al usuario por medio de la funcion `input` que marca de telefono usa, segun la respuesta considere lo siguiente:

- si responde motorola, ofrecerle el modelo motorola M23
- Si responde samsumg, ofrecerle el modelo samsumg A1414
- Si responde ZTE, ofrecerle el modelo ZTE BLADE MAX PRO 2023
- Si responde algo distinto a lo anterior entonces decirle que no conoce dicha marca.

Recuerda que solo puedes usar una vez el codigo `if` y deberas completar el resto con elif y else.


```python
marca_telefono = input("¿Qué marca de teléfono usa? ").lower()

if marca_telefono == "motorola":
    modelo_ofrecido = "Motorola M23"
elif marca_telefono == "samsung":
    modelo_ofrecido = "Samsung A1414"
elif marca_telefono == "zte":
    modelo_ofrecido = "ZTE BLADE MAX PRO 2023"
else:
    modelo_ofrecido = None

if modelo_ofrecido:
    print(f"¡Excelente elección! Le recomendamos el modelo {modelo_ofrecido}.")
else:
    print("Lo siento, no conocemos esa marca de teléfono.")
```
