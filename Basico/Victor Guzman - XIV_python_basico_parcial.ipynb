# Pregunta 1(6 puntos)

Genera un bucle que almacene numeros y los deposite en una lista hasta que los numeros sumen 10.

````python
numeros = []
suma_total = 0

while suma_total < 10:
    numero = float(input("Ingresa un número: "))

    # Añadir el número a la lista
    numeros.append(numero)

    # Actualizar la suma total
    suma_total += numero

print("La suma total es igual o superior a 10.")
print("Lista de números ingresados:", numeros)
2
````

# Pregunta 2(7 puntos)

Genere un programa que le pregunte su nota(numerico) y dependiendo del valor ingresado imprima los siguientes mensajes:
- Si es menor que 11: "Usted desaprobo"
- Si es igual a 11: "Puedes pedir un sustitutorio"
- Si es mayor que 11: "aprobaste"

````python
# Solicitar la nota al usuario
nota = float(input("Ingresa tu nota: "))

# Evaluar la nota e imprimir el mensaje correspondiente
if nota < 11:
    print("Usted desaprobó.")
elif nota == 11:
    print("Puedes pedir un sustitutorio.")
else:
    print("Aprobaste.")
````

# Pregunta 3(7 puntos)

Genera un programa en donde el usuario ingrese su nombre, correo y numero de telefono. Luego de ingresar esos datos mostrar parcialmente los datos ingresados. Por ejemplo:

Estimado Juan, se registro su numero `931*****7` y su correo ser*******@gmail.com

```
# Tiene formato de código
```


Emplee el formato de su preferencia para mostrar parcialmente los datos

````python
# Solicitar datos al usuario
nombre = input("Ingrese su nombre: ")
correo = input("Ingrese su correo: ")
telefono = input("Ingrese su número de teléfono: ")

# Mostrar parcialmente los datos
nombre_par = nombre[0] + "*" * (len(nombre) - 1)
telefono_par = telefono[:3] + "*" * (len(telefono) - 3)
correo_par = correo[:3] + "*" * (correo.index('@') - 3) + correo[correo.index('@'):]

# Imprimir el mensaje con los datos parcialmente ocultos
mensaje = f"Estimado {nombre_par}, se registró su número {telefono_par} y su correo {correo_par}"
print(mensaje)
````
