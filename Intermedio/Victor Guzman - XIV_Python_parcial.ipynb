# Pregunta 1: Calcula la tasa de desempleo por año segun genero (5 puntos)



```python

# Fusionar los datos
data = pd.merge(encuesta, ubigeos, how='left', on='ubigeo')

# Convertir 'ocupado' a tipo numérico
data['ocupado'] = pd.to_numeric(data['ocupado'], errors='coerce')

# Supongamos que desempleo se define cuando 'ocupado' es 0
data['desempleo'] = (data['ocupado'] == 0).astype(int)

# Seleccionar columnas relevantes
subset_desempleo = data[['aÑo', 'mes', 'ubigeo', 'p207', 'desempleo']].copy()
subset_desempleo.rename(columns={'p207': 'gen_pers'}, inplace=True)  # Renombrar 'p207' a 'gen_pers'

# Agrupar por año y género y calcular la tasa de desempleo
grouped_desempleo = subset_desempleo.groupby(['aÑo', 'gen_pers']).agg({'desempleo': ['sum', 'count']})
grouped_desempleo.columns = grouped_desempleo.columns.droplevel()

# Calcular la tasa de desempleo
grouped_desempleo['tasa_desempleo'] = (grouped_desempleo['sum'] / grouped_desempleo['count']) * 100

# Resetear el índice para tener 'año' y 'gen_pers' como columnas regulares
grouped_desempleo.reset_index(inplace=True)

print(grouped_desempleo[['aÑo', 'gen_pers', 'tasa_desempleo']])
```

# Pregunta 2: Define la funcion rango(max - min) y empleala sobre una variable numerica en una tabla pivot cruzando dos campos de tu preferencia (7 puntos)

```python

# Define la función rango
def calcular_rango(columna):
    return columna.max() - columna.min()

# Supongamos que quieres calcular el rango de 'ingreso' cruzando 'aÑo' y 'p207'
tabla_pivot = data.pivot_table(values='ingreso', index='aÑo', columns='p207', aggfunc=calcular_rango)

# Puedes cambiar 'ingreso' por la columna numérica de tu elección
# 'aÑo' y 'p207' son solo ejemplos, puedes cambiarlos según tus necesidades

print(tabla_pivot)
```

# Pregunta 3: Calcula la moda de la edad por año y ocupinf (5 puntos)

```python

# Supongamos que quieres calcular la moda de 'p208a' (edad) cruzando 'aÑo' y 'ocupinf'
tabla_pivot_moda = data.pivot_table(values='p208a', index=['aÑo', 'ocupinf'], aggfunc=lambda x: x.mode().iloc[0])

# Puedes cambiar 'p208a' por la columna de edad o cualquier otra columna cuya moda quieras calcular
# 'aÑo' y 'ocupinf' son los campos por los cuales quieres cruzar los datos

print(tabla_pivot_moda)
```

# Pregunta 4: Exporta las tablas anteriormente calculadas en un solo archivo excel (ver ejemplo en clase) (3 puntos)

```python
# Tabla pivot del rango de ingreso por año y género
tabla_pivot_rango = data.pivot_table(values='ingreso', index='aÑo', columns='p207', aggfunc=lambda x: x.max() - x.min())

# Tabla pivot de la moda de la edad por año y ocupación informal
tabla_pivot_moda = data.pivot_table(values='p208a', index=['aÑo', 'ocupinf'], aggfunc=lambda x: x.mode().iloc[0])

# Nombre del archivo Excel de salida
archivo_excel = 'resultados.xlsx'

# Crear un escritor Excel (pandas ExcelWriter)
with pd.ExcelWriter(archivo_excel, engine='openpyxl') as writer:
    # Guardar cada tabla pivot en una hoja diferente
    tabla_pivot_rango.to_excel(writer, sheet_name='Rango_Ingreso')
    tabla_pivot_moda.to_excel(writer, sheet_name='Moda_Edad')

# Imprimir un mensaje de confirmación
print(f"Las tablas fueron exportadas exitosamente en el archivo: {archivo_excel}")
```
